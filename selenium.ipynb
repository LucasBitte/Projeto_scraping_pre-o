```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Passo 2: Abrir o site
driver.get("https://precodahora.ba.gov.br")  # Substitua pela URL do site

# Passo 3: Esperar até que o botão esteja visível e clicável
botao_tune = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "i.material-icons"))
)

# Passo 4: Clicar no botão
botao_tune.click()

# Esperar alguns segundos para observar a ação
time.sleep(3)

botao_localizacao = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "button.btn-danger.nav-location"))
    )

    # Passo 4: Clicar no botão 'LOCALIZAÇÃO'
botao_localizacao.click()

# Esperar alguns segundos para observar a ação
time.sleep(3)


# Passo 5: Esperar até que o botão 'Usar centro do município' esteja visível e clicável
botao_usar_centro = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "add-center"))
    )
    
# Passo 6: Clicar no botão 'Usar centro do município'
botao_usar_centro.click()

# Esperar alguns segundos para observar a ação
time.sleep(3)

# Passo para encontrar a caixa de pesquisa e inserir texto
campo_pesquisa = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.CLASS_NAME, "sbar-municipio"))
)

# Inserir o texto desejado
campo_pesquisa.send_keys("conqu")  # Substitua "Nome do Município" pelo texto desejado

time.sleep(2)

# Passo para esperar até que o item da lista esteja clicável e então clicar nele
item_lista = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CLASS_NAME, "set-mun"))
)

# Clicar no item da lista
item_lista.click()

time.sleep(2)

# Passo para esperar até que o botão 'APLICAR' esteja clicável e então clicar nele
botao_aplicar = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "aplicar"))
)

# Clicar no botão 'APLICAR'
botao_aplicar.click()


# Passo para esperar até que o campo de entrada esteja clicável e então clicar nele
campo_input = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "fake-sbar"))
)

campo_input.click()
time.sleep(2)
```

### Tela do produto 

```python
# Passo para esperar até que o campo de entrada esteja clicável e então clicar nele
campo_input = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "top-sbar"))  # Localiza o campo de entrada pelo ID
)
# Clicar no campo de entrada
campo_input.click()

# Inserir o texto desejado
campo_input.send_keys("ouro branco")

time.sleep(1)

botao_pesquisa = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "i.fa.fa-search"))  # Localiza o botão pelo seletor CSS
)
# Clicar no botão de pesquisa
botao_pesquisa.click()
```

```python
produtos = ["ouro branco", "redbull", "sonho de valsa", "ferreiro"]

# Loop para iterar sobre cada produto na lista
for produto in produtos:
    # Passo para esperar até que o campo de entrada esteja clicável e então clicar nele
    campo_input = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "top-sbar"))  # Localiza o campo de entrada pelo ID
    )
    
    # Clicar no campo de entrada
    campo_input.click()
    
    campo_input.send_keys(produto)  # Insere o nome do produto atual da lista
    
    # Esperar um segundo (opcional, para evitar problemas de tempo)
    time.sleep(1)
    
    # Localizar o botão de pesquisa e clicar
    botao_pesquisa = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "i.fa.fa-search"))  # Localiza o botão pelo seletor CSS
    )
    botao_pesquisa.click()
    
    # Esperar um tempo para carregar os resultados da pesquisa
    time.sleep(5)  # Ajuste o tempo conforme necessário
    
    # (Opcional) Aqui você pode adicionar lógica para extrair informações dos resultados,
    # como os preços ou nomes dos produtos, antes de passar para o próximo item da lista.
```
